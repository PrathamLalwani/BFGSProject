```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+1.)**2
```

```python
x = -2.
p = -jax.grad(f)(x)
```

```python
phi = lambda alpha: f(x + alpha*p)
```

```python
jax.grad(phi)(0.)
```

```python
# jnp.where(jnp.abs(jax.vmap(jax.grad(phi))(alphas))<1e-2)
```

```python
alphas = jnp.linspace(0.,1.,10000)
c =1e-1
plt.plot(alphas, phi(0)+alphas*c*jax.grad(phi)(0.))
plt.plot(alphas,jax.vmap(jax.grad(phi))(alphas)/jax.grad(phi)(0.))
plt.plot(alphas, f(x+ alphas*p))
plt.show() 
```
